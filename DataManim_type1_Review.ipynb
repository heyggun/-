```python
import datetime

temp = df1[df1['videoname'].str.contains('EP1')].sort_values(by='ct').reset_index(drop=True)
temp['ct'] = pd.to_datetime(temp['ct'])
display(temp.head(3))

idx = temp[(temp['ct'].diff(1) >= datetime.timedelta(minutes=20)) | (temp['ct'].diff(1) <= datetime.timedelta(minutes=5))].index
print(idx)
final_idx = []

for i in idx:
  before = i-1
  after = i+1
  final_idx.append(i-1)
  final_idx.append(i)
  final_idx.append(i+1)

temp[temp.index.isin(final_idx)]
```

Q. **한번씩만 존재하는 (Type1 , Type2)의 쌍을 각 세대(Generation)은 각각 몇개씩 가지고 있는가?**

```python
temp = df[['Type 1','Type 2']].value_counts()
Ans = temp[temp==1].reset_index()

lst = []
for value in Ans.values:
  t1 = value[0]
  t2 = value[1]

  sp = df[(df['Type 1']==t1) & (df['Type 2']==t2)]
  lst.append(sp)

result = pd.concat(lst)
display(result.head(3))

result.reset_index(drop=True)['Generation'].value_counts().sort_index()
```

Q. Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
temp = df.loc[df['Arrival Delay in Minutes'].isnull()][['Class','satisfaction']]
satis = temp[temp['satisfaction']=='satisfied'].groupby('Class').size().reset_index().rename(columns= {0:'satis_cnt'})
non_satis = temp[temp['satisfaction']=='neutral or dissatisfied'].groupby('Class').size().reset_index().rename(columns={0:'non_satis_cnt'})

Ans = pd.merge(satis,non_satis, on='Class')
Ans.loc[Ans['satis_cnt']> Ans['non_satis_cnt']]
```

Q. **흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 charges 평균의 차이는?**

```python
smoker = df[df['smoker']=='yes']['charges'].quantile(0.9)
non_smoker = df[df['smoker']=='no']['charges'].quantile(0.9)

smoker_mean = df[(df['smoker']=='yes') & (df['charges']>=smoker)]['charges'].mean()
non_smoker_mean = df.loc[(df['smoker']=='no') & (df['charges']>=non_smoker)]['charges'].mean()

abs(smoker_mean-non_smoker_mean)
```
