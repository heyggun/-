- x_train: https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv

```python
import pandas as pd
import numpy as np 
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_test.csv ')
```

```python
display(x_train.head(3))
print(x_train.shape)

display(x_test.head(3))
print(x_test.shape)

display(y_train.head(3))
print(y_train.shape)

display(y_test.head(3))
print(y_test.shape)
```

```python
y_train['satisfaction'].value_counts()
```

```python
y_test['satisfaction'].value_counts()
```

```python
y_train_s = y_train.copy()
y_test_s = y_test.copy()

y_train_s['satisfaction'] = y_train_s['satisfaction'].apply(lambda x: 1 if x=='satisfied' else 0)
y_test_s['satisfaction'] = y_test_s['satisfaction'].apply(lambda x: 1 if x=='satisfied' else 0)

print(y_train_s['satisfaction'].value_counts())
print(y_test_s['satisfaction'].value_counts())
```

```python
obj = x_train.select_dtypes(include='object').columns
num = x_train.select_dtypes(include='number').columns

display(x_train[obj].head(3))
display(x_train[num].head(3))

print(obj)
print(num)

print(len(obj)+len(num))
```

```python
print(x_train.isnull().sum())
print(x_test.isnull().sum())
```

```python
x_train['Arrival Delay in Minutes'] = x_train['Arrival Delay in Minutes'].fillna(0)
x_test['Arrival Delay in Minutes'] = x_test['Arrival Delay in Minutes'].fillna(0)
```

```python
def get_values(df):
  cols = df.select_dtypes(include='object').columns

  for col in cols:
    print(df[col].value_counts().index.unique(), len(df[col].value_counts().index.unique()))

get_values(x_train)
get_values(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

def do_Encoding(df):

  cols= ['Gender', 'Customer Type', 'Type of Travel', 'Class']

  for col in cols:
    Encoder = LabelEncoder()
    df[col] = Encoder.fit_transform(df[col])

  return df
```

```python
x_train_l = do_Encoding(x_train)
x_test_l = do_Encoding(x_test)

display(x_train_l.head(3))
display(x_test_l.head(3))
```

```python
x_train['Gate location'].value_counts()
```

```python
x_train['Online boarding'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

def do_Scaler(df):

  cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
  for col in cols:
    Scaler = StandardScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
x_train_s = do_Scaler(x_train_l)
x_test_s = do_Scaler(x_test_l)
```

```python
x_train_s.drop(columns=['ID', 'id'], inplace=True)
x_test_s.drop(columns=['ID','id'], inplace=True)
```

```python
y_train_s
```

```python
from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_train_l,y_train_s['satisfaction'],
                                                            test_size=0.3, random_state=1234)


print(X_train_m.shape)
print(X_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train_m, y_train_m)
prediction = model.predict_proba(X_test_m)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test_m, prediction))
```

```python
model.fit(x_train_l,y_train_s['satisfaction'])
pred = model.predict_proba(x_test_s)[:,1]
```

```python
print(roc_auc_score(y_test_s['satisfaction'], pred))
```

```python
class_pred = model.predict(x_test_s)
```
