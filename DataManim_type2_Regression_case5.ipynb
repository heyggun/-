### 대학원 입학가능성 데이터

- 데이터 설명 : 대학원 입학 가능성 예측 (종속변수 :Chance of Admit)
- x_train: https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/x_train.csv
- y_train: https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/y_train.csv
- x_test: https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/x_test.csv
- x_label(평가용) : https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/y_test.csv

```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/admission/y_test.csv')
```

```python
display(x_train.head(3))
print(x_train.shape)

display(y_train.head(3))
print(y_train.shape)

display(x_test.head(3))
print(x_test.shape)

display(y_test.head(3))
print(y_test.shape)
```

```python
obj = x_train.select_dtypes(include='object').columns
num = x_train.select_dtypes(include='number').columns

print(len(obj)+len(num))
print(len(x_train.columns))

print(obj)
print(num)

display(x_train[obj].head(3))
display(x_train[num].head(3))
```

```python
from sklearn.preprocessing import MinMaxScaler

def do_Scaler(df):

  cols = ['GRE Score', 'TOEFL Score', 'University Rating',
       'SOP', 'LOR', 'CGPA']

  for col in cols:
    Scaler = MinMaxScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
x_train_s = do_Scaler(x_train)
display(x_train_s)

x_test_s = do_Scaler(x_test)
display(x_test_s)
```

```python
from sklearn.model_selection import train_test_split


X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_train_s.drop(columns=['ID','Serial No.']), y_train['Chance of Admit'],
                                                            test_size=0.3,
                                                            random_state=1234)

print(X_train_m.shape)
print(y_train_m.shape)
print(X_test_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
model = RandomForestRegressor()
ada_model = AdaBoostRegressor()
```

```python
model.fit(X_train_m, y_train_m)
prediction = model.predict(X_test_m)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(y_test_m, prediction))
```

```python
ada_model.fit(X_train_m, y_train_m)
ada_prediction = ada_model.predict(X_test_m)

print(r2_score(y_test_m, ada_prediction))
```

```python
model.fit(x_train_s.drop(columns=['ID','Serial No.']), y_train['Chance of Admit'])
pred = model.predict(x_test_s.drop(columns=['ID','Serial No.']))

print('r2 score', r2_score(y_test['Chance of Admit'], pred))
print('rmse', mean_squared_error(y_test['Chance of Admit'], pred, squared=False))
print('mse', mean_squared_error(y_test['Chance of Admit'], pred))
```

```python
ada_model.fit(x_train_s.drop(columns=['ID','Serial No.']), y_train['Chance of Admit'])
ada_pred = model.predict(x_test_s.drop(columns=['ID','Serial No.']))


print('r2 score', r2_score(y_test['Chance of Admit'], ada_pred))
print('rmse', mean_squared_error(y_test['Chance of Admit'], ada_pred, squared=False))
print('mse', mean_squared_error(y_test['Chance of Admit'], ada_pred))
```
