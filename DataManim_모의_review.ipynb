은행 전화 마케팅에 대한 고객 반응 여부


```python
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
```

```python
display(train.head(3))
print(train.shape[0])

display(test.head(3))
print(test.shape[0])
```

```python
train['y'].value_counts()
```

```python
obj = train.select_dtypes(include='object').columns
num = train.select_dtypes(include='number').columns

print(obj)
print(num)

display(train[obj].head(3))
display(train[num].head(3))
```

```python
train.isnull().sum()
```

```python
train['y'] = train['y'].apply(lambda x: 1 if x=='yes' else 0)
```

```python
from sklearn.preprocessing import LabelEncoder

def do_Encoder(df):
  cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
       'month', 'poutcome']

  for col in cols:
    Encoder = LabelEncoder()
    df[col] = Encoder.fit_transform(df[col])

  return df


train_e = do_Encoder(train)
display(train_e.head(3))
```

```python
test_e = do_Encoder(test)
display(test_e.head(3))
```

```python
from imblearn.over_sampling import SMOTE

x_resampled, y_resampled = SMOTE(random_state=0).fit_resample(train_e.drop(columns=['y']),train_e['y'])

y_resampled.value_counts()
```

```python
# 수치형 변수 스케일링
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
Scaler = StandardScaler()
x_resampled[cols] = Scaler.fit_transform(x_resampled[cols])

display(x_resampled)
```

```python
scaled_x_test = train.copy()
display(scaled_x_test.head(3))
```

```python
scaled_x_test[cols] = Scaler.transform(scaled_x_test[cols])
display(scaled_x_test.head(3))
```

```python
# 모델링

from sklearn.model_selection import train_test_split

x_train_m, x_test_m, y_train_m, y_test_m = train_test_split(x_resampled.drop(columns=['ID']), y_resampled,
                                                     test_size=0.3, stratify=y_resampled,
                                                     random_state=2023)

print(x_train_m.shape)
print(x_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train_m, y_train_m)
pred = model.predict_proba(x_test_m)[:,1]
pred_class = model.predict(x_test_m)
```

```python
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score

print(roc_auc_score(y_test_m, pred))
print(f1_score(y_test_m, pred_class))
print(accuracy_score(y_test_m, pred_class))
```
