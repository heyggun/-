Q. 수축기혈압과 이완기 혈압기 수치의 차이를 새로운 컬럼(‘혈압차’) 으로 생성하고, 연령대 코드별 각 그룹 중 ‘혈압차’ 의 분산이 5번째로 큰 연령대 코드를 구하여라

```python
df.columns
```

```python
df['혈압차'] = df['수축기혈압'] - df['이완기혈압']
```

```python
df.groupby(['연령대코드(5세단위)'])['혈압차'].var().sort_values(ascending=False).index[4]
```

Q. 비만도를 나타내는 지표인 WHtR는 허리둘레 / 키로 표현한다. 일반적으로 0.58이상이면 비만으로 분류한다. 데이터중 WHtR 지표상 비만인 인원의 남/여 비율을 구하여라

```python
df['WHtR'] = df['허리둘레'] / df['신장(5Cm단위)']
temp = df.loc[df['WHtR']>= 0.58]
Ans = temp.groupby(['성별코드']).size() 
Ans
```

```python
print(Ans.values[1]/ Ans.values[0])
```

Q. Vehicle_Age 값이 2년 이상인 사람들만 필터링 하고 그중에서
Annual_Premium 값이 전체 데이터의 중간값 이상인 사람들을 찾고, 그들의 Vintage값의 평균을 구하여라

Q. price_range 값이 3인 그룹에서 상관관계가 2번째로 높은 두 컬럼과 그 상관계수를 구하여라

```python
temp = df.loc[df['price_range']==3]
temp = temp.corr().unstack().to_frame().rename(columns={0:'corr'}).dropna()
temp = temp[temp['corr']!=1]
temp.sort_values(by='corr', ascending=False)
```

```python
display(temp.sort_values(by='corr', ascending=False).iloc[2])
display(temp.sort_values(by='corr', ascending=False).iloc[3])
```

데이터 설명 : 비행탑승 경험 만족도 (satisfaction 컬럼 : ‘neutral or dissatisfied’ or satisfied ) (83123, 24) shape

Q. Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
df.columns
```

```python
df['satisfaction'].value_counts()
```

```python
temp = df.loc[df['Arrival Delay in Minutes'].isnull()].groupby(['Class','satisfaction'], as_index=False).size()
temp
```

```python
Ans = temp.pivot(index='Class', columns='satisfaction')
Ans
```

```python
Ans[Ans['size']['neutral or dissatisfied'] < Ans['size']['satisfied']]
```
