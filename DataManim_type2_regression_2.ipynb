```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/X_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/X_test.csv')
x_label = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/y_test.csv')
```

```python
print(x_train.shape)
display(x_train.head(3))

print(y_train.shape)
display(y_train.head(3))


print(x_test.shape)
display(x_test.head(3))

print(x_label.shape)
display(x_label.head(3))
```

`label encoding`

```python
from sklearn.preprocessing import LabelEncoder

def Label_Encoder(df):

  obj = df.select_dtypes(include='object').columns

  for col in obj:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

  return df
```

```python
x_train_l = Label_Encoder(x_train)

display(x_train_l.head(3))
```

```python
x_test_l = Label_Encoder(x_test)
display(x_test_l.head(3))
```

```python
from sklearn.preprocessing import StandardScaler

def Standard_Scaler(df):

  num = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

  for col in num:
    sc = StandardScaler()
    df[col] = sc.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
x_train_s = Standard_Scaler(x_train_l)

display(x_train_s.head(3))
```

```python
x_test_s = Standard_Scaler(x_test_l)
display(x_test_s.head(3))
```

```python
from sklearn.model_selection import train_test_split

x_train_m, x_test_m, y_train_m, y_test_m = train_test_split(x_train_s.drop(columns=['carID']), y_train['price'])

print(x_train_m.shape)
print(x_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train_m, y_train_m)
pred = model.predict(x_test_m)
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test_m, pred))
```

```python
pd.DataFrame({'y_test_m': y_test_m, 'pred':pred})
```

```python
model.fit(x_train_s.drop(columns=['carID']), y_train['price'])
prediction = model.predict(x_test_s.drop(columns=['carID']))

print(r2_score(x_label['price'], prediction))
```

```python
pd.DataFrame({'carID':x_test_s['carID'], 'pred':prediction, 'real':x_label['price']})
```
