```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/x_test.csv')
x_label = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/y_test.csv')
```

```python
print(x_train.shape)
display(x_train.head(3))


print(y_train.shape)
display(y_train.head(3))

print(x_test.shape)
display(x_test.head(3))

print(x_label.shape)
display(x_label.head(3))
```

```python
obj = x_train.select_dtypes(include='object').columns
num = x_train.select_dtypes(include='number').columns

print(obj)
print(num)

display(x_train[obj].head(3))
display(x_train[num].head(3))
```

```python
print(x_train.isnull().sum())
print(x_test.isnull().sum())
```

```python
x_train['region'].value_counts()
```

```python
X_train = pd.get_dummies(x_train)
display(X_train.head(3))

X_test = pd.get_dummies(x_test)
display(X_test.head(3))
```

```python


def do_Scaler(df):

  num = ['age', 'bmi', 'children']
  for col in num:
      sc = StandardScaler()
      df[col] = sc.fit_transform(df[col].values.reshape(-1,1))
  return df

X_train_s = do_Scaler(X_train)
display(X_train_s)
```

```python
X_test_s = do_Scaler(X_test)
display(X_test_s)
```

```python

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_train_s, y_train['charges'])

print(X_train_m.shape)
print(X_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
model = RandomForestRegressor()
model.fit(X_train_m, y_train_m)
prediction = model.predict(X_test_m)
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test_m, prediction))
```

```python
model.fit(X_train_s.drop(columns=['ID']), y_train['charges'])
pred = model.predict(X_test_s.drop(columns=['ID']))
```

```python
print(r2_score(x_label['charges'], pred))
```
