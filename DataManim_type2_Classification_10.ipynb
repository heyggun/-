```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/muscle/y_test.csv')

```

```python
display(x_train.head(3))
print(x_train.shape)
display(y_train.head(3))
print(y_train.shape)

display(x_test.head(3))
print(x_test.shape)
display(y_test.head(3))
print(y_test.shape)
```

```python
print(len(x_train.columns))
print(len(x_train.select_dtypes(include='number').columns))
```

(1) scaling 없이 modeling

```python
X_tr = x_train.drop(columns=['ID'])
y_tr = y_train.drop(columns=['ID'])

X_te = x_test.drop(columns=['ID'])
y_te = y_test.drop(columns=['ID'])
```

```python
from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_tr, y_tr,
                                                            test_size=0.3,
                                                            random_state=2023)

print(X_train_m.shape)
print(y_train_m.shape)

print(X_test_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train_m, y_train_m)
pred = model.predict_proba(X_test_m)[:,1]
pred_class = model.predict(X_test_m)
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
```

```python
print(roc_auc_score(y_test_m, pred))
print(accuracy_score(y_test_m, pred_class))
print(f1_score(y_test_m, pred_class))
```

```python
pred = model.predict_proba(X_te)[:,1]
pred_class = model.predict(X_te)

print(roc_auc_score(y_te, pred))
print(accuracy_score(y_te, pred_class))
print(f1_score(y_te, pred_class))
```
