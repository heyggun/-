```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/hyundai/y_test.csv')
```

```python
display(x_train.head(3))
print(x_train.shape)

display(x_test.head(3))
print(x_test.shape)

display(y_train.head(3))
print(y_train.shape)

display(y_test.head(3))
print(y_test.shape)
```

```python
obj = x_train.select_dtypes(include='object').columns
num = x_train.select_dtypes(include='number').columns

print(obj)
print(num)


display(x_train[obj].head(3))
display(x_train[num].head(3))
```

```python

def get_values(df):
  for col in obj:
    print(df[col].value_counts().index.unique(), len(df[col].value_counts().index.unique()))
  return None
```

```python
get_values(x_train)
```

```python
get_values(x_test)
```

```python
temp = pd.concat([x_train.assign(ind='train'), x_test.assign(ind='test')])
display(temp.head(3))
```

```python
obj = temp.select_dtypes(include='object').columns
num = temp.select_dtypes(include='number').columns

print(obj)
print(num)


display(temp[obj].head(3))
display(temp[num].head(3))
```

```python
temp.isnull().sum().to_frame()
```

```python
temp.describe()
```

```python
temp = pd.get_dummies(temp)
display(temp)
```

```python
from sklearn.preprocessing import StandardScaler

def do_Scaler(df):

  cols = ['year', 'mileage', 'tax(£)', 'mpg', 'engineSize']

  for col in cols:
    Scaler = StandardScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
temp = do_Scaler(temp)
```

```python
x_train_s = temp.iloc[:3888]
x_test_s = temp.iloc[3888:]
```

```python
from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_train_s.drop(columns=['ID']), y_train['price'])

print(X_train_m.shape)
print(X_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train_m, y_train_m)
prediction = model.predict(X_test_m)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(y_test_m, prediction))
print(mean_squared_error(y_test_m, prediction, squared=False))
```

```python
model.fit(x_train_s.drop(columns=['ID']), y_train['price'])
pred = model.predict(x_test_s.drop(columns=['ID']))

print(r2_score(y_test['price'], pred))
print(mean_squared_error(y_test['price'],pred, squared=False))
```
