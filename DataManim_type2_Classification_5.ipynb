```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/insurance/y_test.csv')
```

- x_train dataset에서 모든 컬럼이 결측치인 id 컬럼 제거

```python
X_train = x_train.drop(columns=['id'])
X_test = x_test.drop(columns=['id'])
```

'Previously_Insured' , 'Driving_License' 은 label_encoding 형태이므로,
수치형 전처리 때 제외하기

'Region_Code' 지역코드가 연속형 변수로 파악하기에는 어려우므로, labeling 하기에는 변수가 너무 많아서 일단 제거하고 수행해봄

```python
X_train.drop(columns=['Region_Code'], inplace=True)
X_test.drop(columns=['Region_Code'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

def do_Label_Encoder(df):

  cols = ['Gender', 'Vehicle_Age', 'Vehicle_Damage']

  for col in cols:
    Encoder = LabelEncoder()
    df[col] = Encoder.fit_transform(df[col])

  return df


x_train_l = do_Label_Encoder(X_train)
x_test_l = do_Label_Encoder(X_test)


display(x_train_l.head(3))
display(x_test_l.head(3))
```

```python
from sklearn.preprocessing import StandardScaler

def do_Standard_Scaler(df):

  cols = ['Age', 'Previously_Insured',
       'Annual_Premium', 'Policy_Sales_Channel', 'Vintage']


  for col in cols:
    Scaler = StandardScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df


x_train_s = do_Standard_Scaler(x_train_l)
x_test_s = do_Standard_Scaler(x_test_l)

display(x_train_s.head(3))
display(x_test_s.head(3))
```

```python
from sklearn.model_selection import train_test_split

x_train_m, x_test_m, y_train_m, y_test_m = train_test_split(x_train_s.drop(columns=['ID']), y_train['Response'],
                                                            test_size=0.3, random_state=123)

print(x_train_m.shape)
print(x_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model =RandomForestClassifier()
model.fit(x_train_m, y_train_m)
pred = model.predict_proba(x_test_m)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test_m, pred))
```

```python
model.fit(x_train_s.drop(columns=['ID']), y_train['Response'])
prediction = model.predict_proba(x_test_s.drop(columns=['ID']))[:,1]

print(roc_auc_score(y_test['Response'],prediction))
```

```python
class_prediction = model.predict(x_test_s.drop(columns=['ID']))
```

```python
temp = pd.DataFrame({'ID': x_test_s['ID'], 'pred': class_prediction, 'real': y_test['Response']})
temp
```

```python
RandomForst_wrong = temp.loc[temp['pred']!= temp['real']]['ID'].tolist()
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(n_estimators=200, random_state=1234, learning_rate=0.1)
ada.fit(x_train_s.drop(columns=['ID']), y_train['Response'])
ada_pred = ada.predict_proba(x_test_s.drop(columns=['ID']))[:,1]
```

```python
print(roc_auc_score(y_test['Response'], ada_pred))
```

```python
ada_class = ada.predict(x_test_s.drop(columns=['ID']))

temp2 = pd.DataFrame({'ID': x_test_s['ID'], 'pred' : ada_class, 'real': y_test['Response']})
temp2
```
