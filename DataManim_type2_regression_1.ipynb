```python
from sklearn.preprocessing import LabelEncoder


def Label_Encoder(df):

  obj = df.select_dtypes(include='object').columns

  for col in obj:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

  return df
```

```python
x_train_l = Label_Encoder(x_train)
x_test_l = Label_Encoder(x_test)
```

```python
display(x_train_l.head(3))
display(x_test_l.head(3))
```

```python
# 수치형

from sklearn.preprocessing import StandardScaler

def Standard_Scaler(df):

  num = df.select_dtypes(include='number').columns

  for col in num:
    Scaler = StandardScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
x_train_s = Standard_Scaler(x_train_l)
x_test_s = Standard_Scaler(x_test_l)
```

```python
x_train_s = x_train_s.drop(columns=['StudentID'])
x_test_s = x_test_s.drop(columns=['StudentID'])
```

```python
# 모델링


from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_train_s, y_train['G3'],
                                                            test_size=0.3,
                                                            random_state=1234)

print(X_train_m.shape)
print(y_train_m.shape)
print(X_test_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor


model = RandomForestRegressor()
model.fit(X_train_m, y_train_m)
pred = model.predict(X_test_m)
```

```python
from sklearn.metrics import r2_score, mean_squared_error


print(r2_score(y_test_m, pred))
```
