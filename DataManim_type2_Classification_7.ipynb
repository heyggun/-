```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_train.csv')

x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/waters/y_test.csv')
```

```python
temp = pd.concat([x_train.assign(ind='train'), x_test.assign(ind='test')])
display(temp.head(3))
```

```python
temp['ph'] = temp['ph'].fillna(temp['ph'].mean())
temp['Sulfate'] = temp['Sulfate'].fillna(temp['Sulfate'].mean())
temp['Trihalomethanes'] = temp['Trihalomethanes'].fillna(temp['Trihalomethanes'].mean())
```

```python
x_train = temp[temp['ind']=='train'].drop(columns=['ind'])
x_test = temp[temp['ind']=='test'].drop(columns=['ind'])
```

```python
# 결측치 채운후 재확인

print(x_train.isnull().sum())
print(x_test.isnull().sum())
```

```python
# 수치형 변수 전처리

from sklearn.preprocessing import StandardScaler

def do_Scaler(df):
  
  cols= ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate',
       'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']

  for col in cols:
    Scaler = StandardScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df    
```

```python
x_train_s = do_Scaler(x_train)
display(x_train_s.head(3))

x_test_s = do_Scaler(x_test)
display(x_test_s.head(3))
```

```python
from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_train_s.drop(columns=['ID']), y_train['Potability'],
                                                            test_size=0.3, random_state=2023)

print(X_train_m.shape)
print(X_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train_m, y_train_m)
pred = model.predict_proba(X_test_m)[:,1]
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(y_test_m, pred))
```
