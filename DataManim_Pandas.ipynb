1-57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

Unknown : N Less than 40𝐾:𝑎40K - 60𝐾:𝑏60K - 80𝐾:𝑐80K - 120𝐾:𝑑120K +’ : e

```python
df['Income_Category'].value_counts()
```

```python
Income_dict = {'Unknown' : 'N', 'Less than $40K':'a', '$40K - $60K':'b',
               '$60K - $80K': 'c', '$80K - $120K':'d', '$120K +' : 'e'}
```

```python
df['newIncome'] = df['Income_Category'].map(lambda x: Income_dict[x])
df['newIncome'].head(3)
```

1-58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 

Unknown : N Less than 40𝐾:𝑎40K - 60𝐾:𝑏60K - 80𝐾:𝑐80K - 120𝐾:𝑑120K +’ : e

```python
def get_new_category(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
```

```python
df['newIncome'] = df['Income_Category'].apply(get_new_category)
df['newIncome'].head(3)
```

1-67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당 경우에의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
def fix_year(x):
  import datetime

  year = x.year-100 if x.year >= 2061 else x.year
  return pd.to_datetime(datetime.date(year, x.month, x.day))
```

```python
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_year)
df.head(3)
```

1-76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라.
서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
df.columns
```

```python
def change_date(x):
  import datetime
  date = x.split(':')[0]
  hour = x.split(':')[1]
  
  if hour =='24':
    hour = '00:00:00'

    FinalDate = pd.to_datetime(date + ' ' + hour) + datetime.timedelta(days=1)
  
  else:
    hour = hour + ':00:00'
    FinalDate = pd.to_datetime(date + ' '+hour)

  return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head(3)
```
