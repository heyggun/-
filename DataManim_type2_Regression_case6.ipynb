```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/x_train.csv')
y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/y_train.csv')
x_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/x_test.csv')
y_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/redwine/y_test.csv')
```

```python
display(x_train.head(3))
print(x_train.shape)

display(x_test.head(3))
print(x_test.shape)

display(y_train.head(3))
print(y_train.shape)

display(y_test.head(3))
print(y_test.shape)
```

```python
y_train['quality'].value_counts()
```

```python
obj = x_train.select_dtypes(include='object').columns
num = x_test.select_dtypes(include='number').columns

print(obj)
print(num)

print(len(num))

display(x_train[obj].head(3))
display(x_train[num].head(3))
```

```python
for col in num:
  display(x_train[col].describe())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def do_Scaler(df):
  cols = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

  for col in cols:
    Scaler = MinMaxScaler()
    df[col] = Scaler.fit_transform(df[col].values.reshape(-1,1))

  return df
```

```python
x_train_s = do_Scaler(x_train)
display(x_train_s.head(3))

x_test_s = do_Scaler(x_test)
display(x_test_s.head(3))
```

```python
from sklearn.model_selection import train_test_split

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(x_test_s.drop(columns=['ID']), y_test['quality'],
                                                            test_size=0.3)

print(X_train_m.shape)
print(X_test_m.shape)
print(y_train_m.shape)
print(y_test_m.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
model = RandomForestRegressor()
ada_model = AdaBoostRegressor()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
model.fit(X_train_m, y_train_m)
prediction = model.predict(X_test_m)

print(r2_score(y_test_m, prediction))
```
